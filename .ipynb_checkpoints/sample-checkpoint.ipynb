```python
import random
from datetime import datetime
import re

rental_tuple = []

def has_none_values(parameter):
    return len(parameter) == 0

def in_valid_customer_id(parameter):
    return len(str(parameter)) != 4

def is_Invalid_dates(rental_date,return_date):
    if has_none_values(return_date):
        return False
    rental_date_obj = datetime.strptime(rental_date, '%d-%m-%Y')
    return_date_obj = datetime.strptime(return_date, '%d-%m-%Y')
    return rental_date_obj > return_date_obj

def check_date_format(date_str):
    try:
        date_obj = datetime.strptime(date_str, '%d-%m-%Y')
        date_str = date_obj.strftime('%d-%m-%Y')
        return date_str
    except ValueError:
        return None

def correct_date_format(actual_date):
    date_valid = True
    correction = False
    
    result = check_date_format(actual_date)
    
    if result is None:
        actual_date = re.sub(r'\D', '-', actual_date)
        result = check_date_format(actual_date)
        if result is None:
            date_valid = False
            return date_valid,actual_date,correction
        else:
            correction = True
            return date_valid,actual_date,correction
    return date_valid,actual_date,correction

    
    

games_tuple_list = []
rental_tuple_list = []
suggestion_tuple_list = []
error_list = []


def generate_table(error_records):
    database_initialization = "<h2 style='text-align: center;width: 100%;' > Database Initialization</h2>"
    table_html = database_initialization + "<table style='text-align: start;width: 100%;'>"
    correctly = 'correctly'
    for index, message in enumerate(error_records[:]):
        message = f"{index+1}. " + message
        error_color = 'lightgreen' if correctly in message else 'lightcoral'
        row_html = f"<tr style='background-color: {error_color}'><td>{message}</td></tr>"
        table_html += row_html
    table_html += "</table>"
    return table_html





def data_cleaning(record,index):
    record = record.split(',')
    game_id, rental_date, return_date, customer_id = record
    message = ''
    status = False
    
    if(has_none_values(game_id)):
        message = f"Row={index} has Game Id as Null"
        status = True
        return message,status,record
    
    if(has_none_values(rental_date)):
        message = f"Row={index} has Rental Date as Null"
        status = True
        return message,status,record
    
    if(has_none_values(customer_id)):
        message = f"Row={index} has Customer Id as Null"
        status = True
        return message,status,record
    
    if(in_valid_customer_id(customer_id)):
        message = f"Row={index} has invalid Customer Id of {customer_id} not in range (1000-9999)"
        status = True
        return message,status,record
    
    
    isValid_rental, date_rental, correction = correct_date_format(rental_date)
    
    if(not isValid_rental):
        message = f"Row={index} has Rental Date is invalid, and could not be corrected"
        status = True
        return message,status,record
    else:
        if(correction):
            message = f"Row={index} has Rental Date:  {rental_date}  invalid, but formated correctly to {date_rental}"
        rental_date = date_rental
        record[1]= date_rental
    
    isValid_return, date_return, correction = correct_date_format(return_date)
    
    if(not has_none_values(return_date) and not isValid_return):
        message = f"Row={index} has Return Date is invalid, and could not be corrected"
        status = True
        return message,status,record
    else:
        if(correction):
            if(len(message) > 0):
                message = message + '\n'
            message = message + f"Row={index} has Return Date:  {return_date}  invalid, but formated correctly to  {date_return}"
        return_date = date_return
        record[2]= date_return
    
    if(is_Invalid_dates(rental_date, return_date)):
        message = f"Row={index} has invalid dates, Rental Date is more than Return Date"
        status = True
        return message,status,record
    
    return message,status,record


with open('./data/txt/rental.txt', 'r') as file:
    next(file)
    lines = file.read().splitlines()
    index = 1
    for line in lines:
        code,status,data = data_cleaning(line, index)
        if(status):
            code = code.split('\n')
            error_list = error_list + code
        else:
            if(len(code) > 0 ):
                code = code.split('\n')
                error_list = error_list + code
            rental_tuple_list.append(tuple((data) + [random.randint(0, 15)]))
        index = index + 1 


html_table = widgets.HTML(value=generate_table(error_list))
display(html_table)
        
    
 
```

```python
# Sample data of available titles, genres, and their associated costs and expected sales
inventory = [
    {'title': 'Title A', 'genre': 'Action', 'cost_per_copy': 100, 'expected_sales': 50},
    {'title': 'Title B', 'genre': 'RPG', 'cost_per_copy': 15, 'expected_sales': 30},
    {'title': 'Title C', 'genre': 'Action', 'cost_per_copy': 12, 'expected_sales': 40},
    {'title': 'Title D', 'genre': 'Simulation', 'cost_per_copy': 8, 'expected_sales': 60},
]

# Total budget available for purchasing copies
budget = 1000

# Dictionary to store the purchase recommendations
purchase_recommendations = {}

# Calculate how many copies of each title to purchase based on the budget
for item in inventory:
    title = item['title']
    cost_per_copy = item['cost_per_copy']
    expected_sales = item['expected_sales']

    # Calculate how many copies can be purchased within the budget
    copies_to_purchase = min(budget // cost_per_copy, expected_sales)

    if copies_to_purchase > 0:
        purchase_recommendations[title] = copies_to_purchase

# Display purchase recommendations
for title, copies in purchase_recommendations.items():
    print(f"Purchase {copies} copies of '{title}'")

```

```python
# menu.ipynb

import ipywidgets as widgets
from IPython.display import display

# Assume you have already defined the Game class, game_database, and related functions

import sqlite3
from datetime import datetime

# Update the file paths
game_info_file_path = '/Users/ojeshwarsharma/Desktop/Untitled Folder 1/game_info1.txt'
rental_history_file_path = '/Users/ojeshwarsharma/Desktop/Untitled Folder 1/rental_history1.txt'

# Function to read game information from the file
def read_game_info(file_path):
    games = []
    with open(file_path, 'r') as file:
        next(file)
        next(file)
        for line in file:
            game_info = map(str.strip, line.split(','))
            games.append(tuple(game_info))
    return games

# Connect to SQLite database
connection = sqlite3.connect('ojeshwar.db')
cursor = connection.cursor()

# Create Games table and insert data
with open(game_info_file_path, 'r') as file:
    cursor.execute("DROP TABLE IF EXISTS Games")
    cursor.execute('CREATE TABLE Games (ID INTEGER PRIMARY KEY, Platform TEXT, Title TEXT, Genre TEXT, Purchase_Price REAL, Purchase_Date DATE)')
    for game_info in read_game_info(game_info_file_path):
        cursor.execute("INSERT INTO Games (ID, Platform, Title, Genre, Purchase_Price, Purchase_Date) VALUES (?,?,?,?,?,?)", game_info)
    x = cursor.execute('SELECT * FROM Games')

# Commit changes and close the connection
connection.commit()
connection.close()

action = 0
# Function to handle menu option selection
def handle_menu_selection(change):
    global action
    action = menu_options.value



# Function to display game search results
def display_game_results(matching_games):
    if matching_games:
        for game in matching_games:
            display_game_info(game)
    else:
        print("No games found.")

# Function to display game information
def display_game_info(game):
    print(f"ID: {game.game_id}")
    print(f"Title: {game.title}")
    print(f"Platform: {game.platform}")
    print(f"Genre: {game.genre}")
    print(f"Purchase Price: ${game.purchase_price}")
    print(f"Purchase Date: {game.purchase_date}")
    print(f"Status: {'Rented' if game.is_rented else 'Available'}")
    print("\n")

# Function to handle button click
def handle_search_button_click(button):
    global action
    title = input_text.value
    dropdwon = action 
    with output_area:
        output_area.clear_output()

        if action == 1:
            title = input_text.value
            matching_games = search_game_by_title(title)
            display_game_results(matching_games)

        elif action == 2:
            platform = input_text.value
            matching_games = search_game_by_platform(platform)
            display_game_results(matching_games)

        elif action == 3:
            game_id = int(input_text.value)
            result = rent_game(game_id)
            print(result)

        elif action == 4:
            print('return game')
            title = len(input_text.value)
            if title > 0:
                print(input_text.value)
                game_id = int(input_text.value)
                result = return_game(game_id)
                print('return game')
                print(result)

        elif action == 5:
            display_all_games()

        elif action == 6:
            print("Thank you for using the Video Game Rental Management System. Goodbye!")
            
    matching_games = search_game_by_title(title)
    display_game_results(matching_games)

# Define widgets
menu_title = widgets.HTML("<h1>Video Game Rental Management System</h1>")
menu_options = widgets.Dropdown(
    options=[
        ('Search game by title', 1),
        ('Search game by platform', 2),
        ('Rent a game', 3),
        ('Return a game', 4),
        ('Display all games', 5),
        ('Exit', 6),
    ],
    description='Select Action:',
)

input_text = widgets.Text(description='Enter title/platform/ID:')
search_button = widgets.Button(description='Search by Title')
output_area = widgets.Output()

# Update the menu selection callback
menu_options.observe(handle_menu_selection, names='value')

# Update the button click callback
search_button.on_click(handle_search_button_click)

# Display widgets
display(menu_title)
display(menu_options)
display(input_text)
display(search_button)
display(output_area)
```
